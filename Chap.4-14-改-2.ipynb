```python
from qiskit.algorithms import EstimationProblem
from qiskit.algorithms.amplitude_estimators import AmplitudeEstimation
from qiskit.primitives import Sampler  
from qiskit.circuit import QuantumCircuit
import numpy as np

# 推定したい振幅：sin^2(π/6) = 0.25
theta = np.pi / 6

# A：初期状態を作るユニタリ（1量子ビットで十分な例）
state_prep = QuantumCircuit(1)
state_prep.ry(2 * theta, 0)

# Good状態：測定結果が |1> のときとする
objective_qubits = [0]

# 推定問題の定義
problem = EstimationProblem(
    state_preparation=state_prep,
    objective_qubits=objective_qubits
)

# 振幅推定アルゴリズムの定義（評価用qubit数は精度を決める）
ae = AmplitudeEstimation(num_eval_qubits=3, sampler=Sampler())

# 推定を実行
result = ae.estimate(problem)

# 結果表示
print("推定された振幅:", result.estimation)
print("理論値:", np.sin(theta)**2)
```
